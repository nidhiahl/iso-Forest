## Data Visualisation

```python
fig=plt.figure(figsize=(6,6))
fig.add_subplot(111)
plt.plot(data[:,0],data[:,1],'o', markersize=1, color=[0.5,0.5,0.5])
plt.grid("off")
plt.axis("equal")
plt.xlim([0,50])
plt.ylim([0,50])
plt.tick_params(direction='out', length=6, width=0, colors='k',labelsize=15)
plt.show()
```
