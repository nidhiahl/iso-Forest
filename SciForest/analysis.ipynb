### AUC computation

```python
#Input: read anoamly scores from file.
#output: the computed AUC score.

import pandas as pd
from google.colab import files
import numpy as np
```

```python
files.upload()
```

```python
data = pd.read_csv("Sinusoidal_tested_over_Sinusoidal_exlevel_0.csv", delimiter=' ')

ascores = data['Ascore'].to_numpy()
labels = data['actuallabel']

suspected_anomalies=np.flip(np.argsort(ascores))
print(suspected_anomalies)
```

```python
na = 0
nn = 0
s=0

for rank in range(0, len(suspected_anomalies)):
  # print(ascores[suspected_anomalies[rank]])
  if labels[suspected_anomalies[rank]] == 1:
    s+=rank+1
    na+=1
  else :
    nn+=1

print(na)
print(nn)
print(s)
```

```python
AUC = (s-(na*(na+1))/2 )/(na*nn)
print(AUC)
```
